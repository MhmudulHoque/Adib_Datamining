```python
import pandas as pd
df = pd.read_csv("data/ballers.csv")
print(df)
```

```python
df1=df.drop(["Name","Type","Country","Wickets","Overs","Economy","Strike Rate","Wk/over","Ov/Mtch","Balls","Top","Middle","Lower","CBR","Average"], axis=1)
print(df1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(df1.Matches,df1.Wkpermatch)
plt.xlabel('Matches')
plt.ylabel('Wickets per Match')
plt.show()
```

```python
import numpy as np
a=np.cov(df1.Matches,df1.Wkpermatch)
print("covariance:")
print(a[0,1])
if a[0,1] < 0:
 print("covariance is negative so change is inverse")
elif a[0,1] > 0:
 print("covariance is positive so change is direct")
else:
 print("covariance is 0 so no change")
def pearson(x,y):
 correlation_matrix=np.corrcoef(x,y)
 return correlation_matrix[0,1]
r = pearson(df1.Matches,df1.Wkpermatch)
print("correlation: ")
print(r)
if r < 0 and r <= 0.5:
 print("correlation is negative so relationship is inverse")
elif r < 0 and r > -0.5:
 print("correlation is weakly negative so relationship is weakly inverse") 
elif r > 0 and r >= 0.5:
 print("correlation is positive so relationship is direct")
elif r > 0 and r < 0.5:
 print("correlation is weakly positive so relationship is weakly direct")
else:
 print("correlation is 0 so no relationship")
```
